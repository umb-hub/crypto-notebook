# Secure Hash Algorithm

SHA was originally designed by NIST & NSA in 1993, to be revised in 1996 as SHA-1. SHA is based on design of MD4 with key differences, producing a 160-bit. Today SHA-1 is considered obsolete.

In 2002 NIST adds 3 additional versions of SHA:

- SHA-256
- SHA-384
- SHA-512

Designed for compatibility with AES cipher, increasing security of hash function.



## SHA Versions

In this table are reported SHA version comparision:

| Version | SHA-1 | SHA-224 | SHA-256 | SHA-384 | SHA-512 |
| :- | :- | :- | :- | :- | :- |
| Message digest size | 160 | 224 | 256 | 384 | 512 |
| Message size | $\lt 2^{64}$ | $\lt 2^{64}$ | $\lt 2^{64}$ | $\lt 2^{128}$ | $\lt 2^{128}$ |
| Block size | 512 | 512 | 512 | 1024 | 1024 |
| Word size | 32 | 32 | 32 | 64 | 64 |
| # of steps | 80 | 64 | 64 | 80 | 80 |

## SHA Algorithm

Now examine the structure of SHA-512, noting that the other versions are quite similar.

- Append padding bits, consists of a single 1-bit followed by the necessary number of 0-bits, so that its length is in modulo 1024.
- Append length as an (big-endian) unsigned 128-bit integer
- Process the message in $N$ 1024-bit (128-word) blocks. Each round takes in input message block into a cryptographic function $F$ and xor result with a 512-bit buffer $H_i$ ($H_0$ is initialized by an inizitaliation vector IV). Output is an 512-bit buffer $H_{i+1}$ used in next round.
- Output the final state value $H_N$ as the resulting hash.

Cryptographic Function $F$ is also an compression function that maps message (1024-bits) in 512-bits

## SHA Security

| Version | Status |
| :- | :- |
| SHA-1 | It is considered **broken**. |
| SHA-2 | In use but similar to SHA-1 |
| SHA-3 | In use and devolped to be more secure |

On 23 February 2017, Google announced the first SHA-1 public collision using the SHAttered attack (100k faster than the brute force attack). The collision was demonstrated on two different PDF files. Therefore it is considered broken.

```python
import hashlib
from binascii import hexlify
message = b"My message"

print("SHA-1    %s" % hexlify(hashlib.sha1(message).digest()))
print("SHA-224  %s" % hexlify(hashlib.sha224(message).digest()))
print("SHA-256  %s" % hexlify(hashlib.sha256(message).digest()))
print("SHA3-256 %s" % hexlify(hashlib.sha3_256(message).digest()))
print("SHA-512  %s" % hexlify(hashlib.sha512(message).digest()))
print("SHA3-512 %s" % hexlify(hashlib.sha3_512(message).digest()))
```
