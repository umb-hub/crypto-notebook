# Block Cipher

A block cipher is a deterministic cipher, in which plaintext and ciphertext are same finite set.

An element of finite set is lenght-fixed and it is colled **data-block**.

## Block Cipher Principles

Block Cipher should looked as a random permutation on plaintext to get ciphertext, i.e. a secure block cipher should be *computationally indistinguishable* from a random permutaton. 

Block cipher must be able to decrypt ciphertext effienctly, so permutation must be a invertible function. In order to define an invertible block encryption, exists some known structures: most symmetric block ciphers are based on **Feistel Cipher Structure**

## Shannon Cipher

Shannon introduced idea of substitution-permutation (**S-P**) network, that forms basis of modern block cipher.

S-P networks are based on two primitive cryptographic operations:

- Substitution (**S-Box**): provide to *confusion* of key
- Permutation (**P-Box**): provide to *diffusion* of plaintext

Throught S-P networks, cipher needs to obscure statical properties of original messages:

- Confusion: make relationship between ciphertext and key as complex as possible
- Diffusion: dissipates statistical structure of plaintext in ciphertext

Note: one-time pad is a shannon cipher perfectly secure, because it guarantees maximum diffusion of plaintext.

## Block Cipher Design

In pratice, all modern block ciphers are designed as **iterated ciphers**, that include 2 components:

- **Key Scheduling**: produce a set of subkeys from a masterkey
- **Round Function**: is iterated on message for each subkey

Block Cipher Design

Is iteration cipher the right way to implements confusion and diffusion ? **Nobody Knows!**

Heuristically, iterating is a simple way to makes a cipher secure but **not any function is good for iteration**.

In general, analyzing security of block cipher is considered a hard problem.

## Feistel Cipher Structure

Based on concept of invertible product cipher and partitionating input blocks into two halves.

It implements Shannon's S-P network concept with an iterative process.

Feistel Structure

Steps in a round:
- Input blocks is divided in 2 halfs
- Output left part block is input right part block
- Output right part block is round function to input right part block and xor with left part block

At the end of iterative process, half parts of blocks are swapped.

A single round doesn't guarantees confusion and diffusion, but iterating process guarantees these properties more and more to any interaction. So it's important define how many round function apply to plaintext, because too much rounds are useless and computationally expensive.

Feinstel Structure is a reversible structure for anyone round function (i.e. round function could be a no invertible function). 

In order to reverse Feinstel Structure, reapplying same structure with reverse-ordered subkeys, it's possible obtain plaintext from ciphertext. 


### Demonstration reversibility

Reapplying same structure with reverse-ordered subkeys, after first round:

#### Left Part

Feistel First Round

$$ LD_1 = RD_0 = LE_{16} = RE_{15}$$



#### Right Part

Feistel First Round

$$ RD_1 = LD_0 \oplus F(RD_0, K_{16}) = RE_{16} \oplus F(RE_{15}, K_{16}) = [LE_{15} \oplus F(RE_{15}, K_{16})] \oplus F(RE_{15}, K_{16}) = LE_{15} $$



### Feistel Cipher Design

In order to design a cipher on Feinstel Structure, it's needed define some parameters:

- Block Size
- Key Size
- Number of rounds
- Key Scheduling
- Round Function
- Fast hardware/software implementation of en/decryption
- Ease of analysis (Kerckhoff principle)
