# AES

AES (Advanced Encryption Standard) is new block cipher standard proposed to NIST in 1997 by Daemen and Rijmen (*Rijndael*), Belgian cryptographers. AES block cipher operates on 128-bit blocks and support 128, 192 and 256-bits key, in order to be update due to progressive computational power.

AES is an iterate cipher that iterates a simple round cipher several times and *is not based on feinstel structure* and designed to be implemented by software:

- Resistant aginst known attacks
- Speed and code compactness on many CPUs
- Design semplicity



## Iterative Schema

AES proceses data as matrix 4x4 bytes and operates on matrix each round. An Initial Trasformation is applied and later multiple rounds, any round produce an output called **state** and it's input in next round.

AES 10 Round

| Key-Size | # Rounds | 
| :-: | :-:|
| 128 bit | 10 rounds |
| 192 bit | 12 rounds |
| 256 bit | 14 rounds |



## Key Expansion

Key scheduling algorithm has in input a key with any possible lenght (128/192/256 bits), from which generates a 4x4 matrix bytes for each round and initial trasformation.  

## Round Functions

Each round is composed by:

- Byte substitution
- Shift rows
- Mix columns
- Add round key
- Scramble (Introduce non-linearity ?)

In order to decrypt, AES Round Function needs to be invertible.

### Byte substitution

AES use a S-Box of 16x16 bytes. Each byte of state is replaced by byte indexed by row (left 4 bits) and column (right (4-bits))
E.g.

Byte {0x95} = 1001 0101 -> replaced by byte in row 9 (1001) and column 5(0101) = {0x2A}

S-Box is construced on defined trasformation of valued in Galois Field (2^8) and it is designed to be resistent to all known attacks.

### Shift Rows

Each rounds compute a circular byte shift:

- 1st row is unchanged
- 2nd row does 1 byte circular shift to left
- 3rd row does 2 byte circular shift to left
- 4th row does 3 byte circular shift to left

To decrypt: using byte circular shift to right.



### Mix Columns

Each rounds compute a mix of columns, in which **each column** is processed. This operation can be computed by matrix multiplication in $GF(2^8)$, obtained using prime poly:

$$ m(x) = x^8+x^4+x^3+x+1$$

## Key Expansion

AES-128 needs to generate 11 rounds keys where each round is a 4x4 byte matrix, composed by 128-bits. AES-key is partitioned into four 32-bit words and these forms first round key. Each other round key are generated by precedent round key through an algorithm.


immagine.png

Image shows each group of 4 bytes in the key being assigned to the first 4 words, then the calculation of the next 4 words based on the values of the previous 4 words, which is repeated enough times to create all the necessary subkey information.

Key expansion is designed to be resistent to know attacks:

- Invertible trasformation
- No weak keys
- Diffuse key bits into round keys
- Enough non-linearity
- Etc...




## AES Decryption

The AES decryption cipher is not identical to the encryption cipher,sequence of transformations for decryption differs from that for encryption, although the form of the key schedules for encryption and decryption is the same.

This has the disadvantage that two separate software or firmware modules are needed for applications that require both encryption and decryption.



immagine.png

There is, however, an equivalent version of the decryption algorithm that has the same structure as the encryption algorithm, with the same sequence of transformations as the encryption algorithm (with transformations replaced by their inverses). To achieve this equivalence, a change in key schedule is needed. 

By constructing an equivalent inverse cipher with steps in same order as for encryption, we can derive a more efficient implementation, that makes the decryption key schedule a little more complex with this construction, but allows the use of same h/w or s/w for the data en/decrypt computation.
