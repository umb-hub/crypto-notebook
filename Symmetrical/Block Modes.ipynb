# Block Cipher Mode

Block ciphers encrypt only fixed sized block, in order to encrypt an arbitrary amounts of data NIST defines 5 odes, that can be destinguished in **block and stream modes**:

- ECB (Elettronic Code Book) : block mode
- CBC (Cipher Block Chaining) : block mode
- CFB (Cipher FeedBack) : stream mode
- OFB (Output FeedBack) : stream mode
- CTR (Counter) : stream mode



# Block Modes



## Elettronic Code Book (ECB)

Simplest way to encrypt a large block of data is ECB, dividing message in n fixed sized blocks and encrypt them with same algorithm and same key.

ECB - Elettronic Code Book

An attacker can easily detect when two blocks are equals, because they are encrypted in same ciphertext. ECB mode can be osserved as a simple substitution algorithm with a long alphabet, on which can be used all statistical techniques.

Futhermore integrity can be easily violated replacing single block in any position.

ECB Penguin

For those reason ECB is strongly discouraged, but if message is fits into a block o few blocks (can't be applied statistical techniques), ECB can be used.



## Cipher Block Chaining (CBC)

CBC approach is based on linking consecutive block in encryption operation. Each previous ciphertext block is chained with current plaintext block. In order to apply CBC to first block an **Initial Vector** (IV) need to be declared.

$$ C_i = E((P_i \oplus C_{i-1}),K)$$

with

$$ C_{-1} = IV$$

**Note:**

**If a block is lost then message can't be recovered.**

## Message Padding

In block mode if message can't be divided perfectly in n block, last block needs to be padded.

At the end of message are added prefixed and known bytes in order to be un-padded after decryptation.

# Stream Modes

Block modes encrypt entire blocks, but sometimes in realtime data may need to operate in smaller unit. In this case, block cipher are used as PRG (*Pseudo-Random Generator*).



## Cipher FeedBack (CFB)

In stream mode messages must be treated as a stream of bits. Stream cipher use XOR encrypt stream message with a key stream. CFB generate a key stream using block ciphers.

In a s-bit CFB (CFB-s) mode is defined an Initialition Vector (IV) a block cipher encrypt it, in order to generate a initial key-stream. Each iteraction s bits of avaiable message are XOR with s bits of ciphered IV. Next iteration IV is shifted by s bits and last s bits are filled up by ciphered bits, in order to generate a new key-stream.

CFB

Note:

An error propagation can be dangerous for entire message.



## Output FeedBack (OFB)

In OFB mode first key-stream block is generated by block cipher using a **Nonce**, and next block is generated only reusing previous key-stream block. So key-stream is not plaintext/ciphertext dependent and could be calculated knowing *nonce* and *key* of block cipher.

Note:

An error propagation is not so dangerous for entire message.



## Counter (CTR)

In Counter mode a single key-stream block is directly indipendent each other, they are indirectly dependent, because each key-stream block is generated by a key and a counter, that increase with each block.

Counter is set in a random value that will be setted between two peers.
