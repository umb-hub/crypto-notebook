# Hash Functions

Hash function is a public function, that maps a arbitrary message to a fixed size ciphertext.

In general terms, the principal object of a hash function is data integrity. A change to any bit or bits in M results, with high probability, in a change to the hash code.

CRC32 is an easy example of hash functions, useful for data integrity against transmission errors (*random errors*), but it can be easily computed from an adversary that could generate a new message with same hash.

A **cryptographic** hash function is an algorithm for which it is computationally infeasible to find:

- data mapping to a specific hash (*one-way property*)
- two data to same hash (*collision-free property*)

Anyway it is impossible create a perfect **cryptographic hash function** due to fixed size of co-domain.

## Common use of Hash Functions and requirements

Hash functions are used to:

- Crate a one-way password file, storing hash of password and not actual password.
- Intrusion detection and virus detection, storing and checking file's hash on system.
- Pseudo random function (PRF) or pseudorandom number generator(PRNG)

| Requirement | Description |
| :- | :- |
| Variable input size | $H(x)$ can be applied to a block of any size. |
| Fixed output size | $H(x)$ produce a fixed-length output. |
| Efficienty | $H(x)$ is really easy to computed for any given x. |
| Preimage resistant (*One-way property*) | For any given hash value $h$, it is computationally infeasible to find an $y$ such that $H(y) = H(x)$ . |
| Second Preimage resistente (*Weak Collision Resistant*) | For any given block $x$, it is computionally infeasible to find $y$ such that $H(y) = H(x)$ . |
| Collision Resistant (*Strong Collision Resistant* ) | It is computationally infeasible to find any pair $(x,y)$ such that $H(x) = H(y)$ . |
| Pseudorandomness | Output of $H(x)$ meets standard tests for pseudorandomness . |

Weak and Strong Collision Resistant looks similar, but:

- In Weak Collision Resistant: manipulating known input block $x$ it is computationally infeasible to find $y$
- In Strong Collision Resistant: it is impossible find any pair $(x,y)$ that collides, without any restrictions.

## Attacks on Hash Functions

As with encryption algorithms, there are two categories of attacks on hash functions: brute-force attacks and cryptanalysis.

### Brute-force attacks

**Brute-force attack** does not depend on the specific algorithm but depends only on bit length. In the case of a hash function, a brute-force attack depends only on the bit length of the hash value.


#### Preimage or second preimage attack

For a preimage or second preimage attack, an adversary wishes to find a value $y$ such that $H(y)$ is equal to a given hash value $h$. The brute force method is to pick values of $y$ at random and try each value until a collision occurs. 

For an m-bit hash value, the level of effort is proportional to $2^m$. Specifically, the adversary would have to try, on average, $2^{mâ€“1}$ values of $y$ to find one that generates a given hash value $h$. 
 
#### Collision resistence attack

For a collision resistant attack, an adversary wishes to find two messages or data blocks, $x$ and $y$, that yield the same hash function: $H(x) = H(y)$. This requires much less effort than a preimage or second preimage attack.

For an m-bit hash value, the level of effort is proportional to $2^{m/2}$, due to birthday paradox: the chance that in a group of people two will share the same birthday is not low. Infact in a group of 23 people, collision probability is more than 50%.

So we can think a 64-bit hash is secure, but due to birthday paradox it is not, because it needs only $2^{64/2}$ attemps.



### Cryptanalysis Attacks

- **Cryptanalysis attack**, in contrast, is an attack based on weaknesses in a particular cryptographic algorithm.

Hash Function using Block Cipher

Hash fuctions can use an iterative structure, i.e. processing message using cipher block chaining technique, where final block outputs hash value. Resulting hash could be too small due to block ciphertext length, for example DES produce a 64-bit hash. Using cipher block chaining technique, hash functions could be vulnerable to:

- Birthday attack
- Meet-In-The-Middle attack

So cipher block chaining technique is a really delicate approach, that needs to be well built, like in [SHA](./Hash/SHA.ipynb).
